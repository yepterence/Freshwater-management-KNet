```
import os
import geemap
import pandas as pd
import geopandas as gpd
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
from google.colab import drive
```

```
# Access google drive
drive.mount('/content/drive')
```

```
#credentials
api = SentinelAPI(os.getenv('USERNAME'), os.getenv('PASSWORD'))
```

```
# go to http://geojson.io/ and select the polygon of the area. then download the geoJson file of the selected polygon. 
footprint = geojson_to_wkt(read_geojson('/content/map.geojson'))
products = api.query(footprint,
                     date=(date(2022,2,1), date(2022,2,10)),
                     platformname='Sentinel-3',
                     producttype='SR_2_LAN___',
                     timeliness='Non Time Critical',
                     productlevel='L2')

api.download_all(products)
```
